```python
#Importing all the necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("icc_wc_23_bat.csv")
df.head(5)
```

```python
null_values = df.isnull().sum()
print(null_values)
```

```python
print(df.describe())
```

```python
print(df.dtypes)
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(x='team', y='strike_rate', data=df)

plt.title('Team Strike Rate Chart')
plt.xlabel('Team')
plt.ylabel('Strike Rate')
plt.xticks(rotation = 90)

plt.show()


#the financial landscape of cricket teams can evolve due to changing market dynamics, team performance, and the overall popularity of the sport.
#the budget the world cup this year was 84 crores inr or 10 million usd. this was a 40% increase from the previous budget set in 2019 and has contributed to the practice of the players which has significantly improved the run rates of india as we can see in the graph
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(x='team', y='strike_rate', data=df)

plt.title('Team Strike Rate Chart')
plt.xlabel('Team')
plt.ylabel('Strike Rate')
plt.xticks(rotation = 90)

plt.show()
```

```python
rohit_data = df.loc[df['team'] == 'India']

plt.figure(figsize=(10, 6))
sns.scatterplot(x='minutes', y='runs', data=rohit_data)
plt.title('India - Runs vs Minutes')
plt.xlabel('Runs')
plt.ylabel('Minutes')
plt.show()

#india's runs show a general increase in the first 80 minutes. The exemplary performance shown by shubhman gill, 
#suryakumar yadav and such players got india to score 140 runs in a staggering 60 minutes time period and even 
#managed to get us to 190 in 100 minutes
```

```python
team_runs = df.groupby('team')['runs'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='team', y='runs', data=team_runs, palette='viridis')
plt.title('Runs by Team')
plt.xlabel('Teams')
plt.xticks(rotation = 45)
plt.ylabel('Total Runs')
plt.show()

#this colorful plot shows us runs scored by the teams participating in the mens world cup 
#australia gave an exemplary performance, one of the many factors contributing to this win other than the leadership and camrederie were the rich cricket infrastructure and adaptability to the field and conditions of game.
#the total runs scored by australia were under 3000 and india dominated as a powerhouse with the upper hand of over 3000 runs
```

```python
team_boundary_chart = df.groupby('team')[['4s', '6s']].sum().plot(kind='bar', stacked=True)
plt.xticks(rotation = 45)
plt.xlabel('Teams')

#this graph highlights the number of boundaries (4s and 6s) scored by each team.
#by comparing the two dominating teams australia and india, australia has a slight dominance over india in the number of 6s scored. 
#this can be influenced by the pitch conditions during the game ,the batting style of the teams and general game stratergy
#Indian batsmen adopt a low stance and exhibit a driving tendency due to lower bounce on Indian pitches, while Australians stand more erect, favoring staying on top of the bounce and employing backfoot movement. Indian footwork involves a small stride followed by a longer one, while Australians employ a back-and-across movement. Strong zones for Indians are the "V" and leg side, while Australians score more on the square of the wicket.
```

```python
virat_data = df.loc[df['player'] == 'Virat Kohli']
travis_data = df.loc[df['player'] == 'Travis Head']

plt.figure(figsize=(10, 6))

sns.lineplot(x='runs', y='balls', data=virat_data, label='Virat Kohli')
sns.lineplot(x='runs', y='balls', data=travis_data, label='Travis Head')

plt.title('Comparison of Runs: Virat Kohli vs. Travis Head')
plt.xlabel('Runs')
plt.ylabel('Balls')
plt.legend()

plt.show()
```

```python
team_strike_rate_chart = sns.boxplot(x='team', y='strike_rate', data=df)
plt.xticks(rotation = 90)
#generally the higher the strike rate is defined by the average number of runs scored by a batter in the course of 100 balls
#the strike rate of batters this year were lead by glenn maxwell with a 150 followed by henrich klassen and on seventh position rohit sharma with a stat of 125.94
```

```python
virat_data = df.loc[df['player'] == 'Virat Kohli']
plt.figure(figsize=(10, 6))
sns.barplot(x='runs', y='balls', data=virat_data)
plt.title('Virat - Runs vs Minutes')
plt.xlabel('Runs')
plt.ylabel('Minutes')
plt.show()

#for india's favourite player none other than virat kohli the runs generally showed a pattern of increase from the 100th minute with a rise and fall till 120th.
#virat kohli's performance was generally influenced by his form and fitness, role in the match and team dynamics
#Kohli scored a total of 765 runs in 11 innings, the highest-ever runs scored by a batter in a single edition of the World Cup.
```

```python
selected_player = 'Virat Kohli'
selected_player_data = df[df['player'] == selected_player]
plt.figure(figsize=(8, 8))
plt.pie(selected_player_data['strike_rate'], labels=[f"Match {i+1}" 
for i in range(len(selected_player_data))], autopct='%1.1f%%', startangle=90, colors=plt.cm.viridis.colors)
plt.title(f'Strike Rate Distribution for {selected_player}')
plt.show()
```

```python
#Rachin Ravindra- The rising star
selected_player = 'Rachin Ravindra'
filtered_df = df[df['player'] == selected_player]
plt.figure(figsize=(10, 6))
heatmap_data = filtered_df.pivot_table(index='player', columns='minutes', values='runs', aggfunc='sum', fill_value=0)
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title(f'Player Impact in Each Match - {selected_player}')
plt.show()
```
